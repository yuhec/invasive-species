```
import pickle
import pandas as pd
import numpy as np
import sys
import os
sys.path.append('../..')
from invasive_species.src.process_data import process_data
```

```
loaded_model = pickle.load(open('../data/model_sklearn_mlpclassifier.pkl', 'rb'))
```

```
TEST_DIR = '/datasets/louis/test'
IMG_SIZE = 224
```

```
x_test, img_nums = process_data(TEST_DIR, len(os.listdir(TEST_DIR)), IMG_SIZE)
```

```
x_test_np = np.array(x_test)
nb, x, y = x_test_np.shape
x_test_2d = x_test_np.reshape(nb, x*y)
```

```
y_test_pred = loaded_model.predict_proba(x_test_2d)
```

```
y_test_pred = pd.DataFrame(y_test_pred)
y_test_pred['name'] = img_nums
y_test_pred = y_test_pred.rename(columns={0:  'invasive'})
y_test_pred = y_test_pred[['name', 'invasive']]
y_test_pred.to_csv('../data/submission.csv', index=False)
```
